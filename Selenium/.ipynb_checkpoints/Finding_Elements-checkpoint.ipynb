## Find element by XPATH



```python
results = driver.find_elements(By.XPATH, '/html/body/div')
```

```python
results
```

### Найти третий 'div' елемент

```python
results = driver.find_elements(By.XPATH, '/html/body/div[3]')
results
```

```python
results[0].text
```

### Найти все 'div' элементы

```python
results = driver.find_elements(By.XPATH, '//div')
len(results)
```

### Найти все 'div' элементы c 'id' атрибутом = 'bodyContent'

```python
results = driver.find_elements(By.XPATH, "//div[@id='bodyContent']")
len(results)
```

```python
results[0]
```

### Мы можем использовать элемент, который мы нашли, чтобы найти его родителя

```python
parrent = results[0].find_element(By.XPATH, '..')
```

```python
# parrent.get_attribute('innerHTML')
```

Внутри находится элемент "bodyContent"

### Узнать текст по XPATH

```python
results = driver.find_element(By.XPATH, "//span[text()=\"From today\'s featured article\"]")
```

```python
results.text
```

### Можно не писать весь текст, а использовать инструментарий

```python
results = driver.find_element(By.XPATH, "//span[starts-with(text(), 'From')]")
results.text
```

### Или можно использовать функцию 'contains'

```python
results = driver.find_element(By.XPATH, "//span[contains(text(), 'today')]")
results.text
```

### Найти элемент, который в себе что-то содержит

```python
results = driver.find_element(By.XPATH, "//div[contains(@id, 'body')]")
results
```
