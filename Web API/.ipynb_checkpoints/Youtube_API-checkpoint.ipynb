### Adele Vevo Youtube page

```python
url = 'https://www.youtube.com/user/adelelondon/videos'
```

#### Step 1 (get id)

```python
channel_name = 'Adele'
```

```python
url = 'https://www.googleapis.com/youtube/v3/channels'
```

```python
params = {'part': 'snippet',
          'forUsername': channel_name,
          'key': token}
```

```python
response = requests.get(url, params=params)
print(response.status_code)
```

```python
response.json()['items']
```

```python
channel_id = response.json()['items'][0]['id']
```

#### Step 2 (get videos)

```python
url = 'https://www.googleapis.com/youtube/v3/search'
```

```python
params = {'part': 'snippet',
          'channelId': channel_id,
          'key': token,
          'type': 'video',
          'order': 'date',
          'maxResults': 20}
```

```python
response = requests.get(url, params=params)
print(response.status_code)
```
